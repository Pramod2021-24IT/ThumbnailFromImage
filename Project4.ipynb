# Create a Thumbnail from an Image:
   - Read a large image, resize it to create a thumbnail, and save the thumbnail to a file. Display both the original and thumbnail images using Matplotlib.

```python
import cv2
import matplotlib.pyplot as plt
```

```python
# Read the image
image = cv2.imread(r'C:\Users\param\Pictures\Screenshots\Screenshot 2024-10-22 204844.png')

# Convert the image from BGR to RGB color space (OpenCV reads images in BGR by default)
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
# Define the size for the thumbnail
thumbnail_size = (100, 100)  # You can adjust this size as needed

# Resize the image to create a thumbnail
thumbnail = cv2.resize(image_rgb, thumbnail_size, interpolation=cv2.INTER_AREA)
```

```python
# Save the thumbnail image
cv2.imwrite('thumbnail_image.jpg', cv2.cvtColor(thumbnail, cv2.COLOR_RGB2BGR))
```

```python
# Display the images
plt.figure(figsize=(10, 5))

# Display original image
plt.subplot(1, 2, 1)
plt.imshow(image_rgb)
plt.title('Original Image')
plt.axis('off')

# Display thumbnail image
plt.subplot(1, 2, 2)
plt.imshow(thumbnail)
plt.title('Thumbnail Image')
plt.axis('off')

plt.show()
```
